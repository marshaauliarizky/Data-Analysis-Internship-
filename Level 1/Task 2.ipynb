## Task 2: Exploratory Data Analysis (EDA)

#### Summary Statistics

```python
import pandas as pd

df = pd.read_csv("iris.csv")

# Summary statistics
print("Summary Statistics:")
print(df.describe())

# Mean
print("\nMean:")
print(df.mean(numeric_only=True))

# Median
print("\nMedian:")
print(df.median(numeric_only=True))

# Mode
print("\nMode:")
print(df.mode().iloc[0])

# Standard Deviation
print("\nStandard Deviation:")
print(df.std(numeric_only=True))
```

#### Interpretation 

In this step, I calculated the main summary statistics of the dataset, including mean, median, mode, and standard deviation.

These statistics help to:

- Understand the central tendency of each numerical feature.

- Measure data spread and variability.

- Identify potential outliers or unusual patterns.

From the Iris dataset, we can observe:

- The mean and median values are relatively close, indicating balanced distributions.

- Standard deviation shows moderate variability in petal and sepal measurements.

#### Histograms (Data Distribution)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")

# Histograms for numerical columns
df.hist(figsize=(10, 8))
plt.suptitle("Distribution of Numerical Features", fontsize=14)
plt.show()
```

#### Interpretation

Histograms help visualize how the values of each numerical feature are distributed.

From the Iris dataset:

- Sepal and petal measurements show roughly normal distributions.

- Petal length and petal width appear slightly skewed.

- The spread indicates natural biological variation between flower species.

This helps us understand whether the data is normally distributed and if there are extreme values.

#### Boxplots (Detecting Outliers)

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df.select_dtypes(include='number'))
plt.title("Boxplot of Numerical Features")
plt.show()
```

#### Interpretation

Boxplots help detect:

( Median values, Spread of data, Potential outliers )

From the visualization:

- Most features show consistent ranges.

- Petal measurements may show wider variation compared to sepal features.

- No extreme outliers are clearly visible.

This confirms the dataset is relatively clean and well-distributed.

#### Scatter Plot (Feature Relationships)

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x="petal_length",
    y="petal_width",
    hue="species"
)

plt.title("Petal Length vs Petal Width")
plt.show()
```

#### Interpretation

The scatter plot shows a clear separation between species:

- Setosa appears distinctly separated from the other two species.

- Versicolor and Virginica overlap slightly but still form visible clusters.

- Petal length and petal width show a strong positive relationship.

This indicates that petal measurements are strong indicators for species classification.

#### Correlation Matrix

```python
plt.figure(figsize=(8,6))

correlation = df.corr(numeric_only=True)

sns.heatmap(
    correlation,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Matrix")
plt.show()
```

#### Interpretation

The correlation matrix shows:

- Petal length and petal width have a very strong positive correlation.

- Sepal measurements show moderate correlation.

- No strong negative correlations are present.

This suggests:

- Petal features are highly related to each other.

- These features would likely be important predictors in classification models.

#### Final EDA Conclusion 

The exploratory data analysis revealed that petal measurements strongly differentiate species in the dataset. The distributions are relatively balanced, with no extreme outliers. Correlation analysis indicates a strong positive relationship between petal length and petal width, making them important variables for predictive modeling.
