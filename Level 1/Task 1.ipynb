```python
import pandas as pd

df = pd.read_csv("iris.csv")

print(df.head())
print(df.info())
```

#### Identify and Handle Missing Values


```python
# Check missing values
print("Missing values per column:")
print(df.isnull().sum())
```

```python
# Handle missing values (for numeric columns)
df.fillna(df.mean(numeric_only=True), inplace=True)
```

#### Remove Duplicate Rows

```python
print("Rows before removing duplicates:", df.shape[0])

df = df.drop_duplicates()

print("Rows after removing duplicates:", df.shape[0])
```

#### Standardize Inconsistent Data Formats

```python
if 'species' in df.columns:
    df['species'] = df['species'].str.lower().str.strip()

print(df['species'].unique())
```

#### During preprocessing, I validated the dataset for missing values, duplicate records, and inconsistent categorical formats. Although the dataset was relatively clean, I implemented standard preprocessing techniques such as mean imputation, duplicate removal, and categorical standardization to ensure data quality and analytical reliability.
