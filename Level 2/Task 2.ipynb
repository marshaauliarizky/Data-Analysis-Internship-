## Task 2: Time Series Analysis

#### Prepare the Date Column

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Stock Prices Data Set.csv")

# Convert date to datetime
df['date'] = pd.to_datetime(df['date'])

# Sort by date
df = df.sort_values('date')

# Set date as index
df.set_index('date', inplace=True)

print(df.head())
```

#### Plot Time Series

```python
plt.figure(figsize=(12,5))
plt.plot(df['close'])
plt.title("Stock Closing Price Over Time")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.show()
```

#### 
The time-series plot reveals a clear upward trend in the stock’s closing price over time.
While short-term fluctuations are present, the overall direction indicates consistent long-term growth.
The strongest price increase occurs between 2016 and 2017, suggesting a period of significant market expansion.

```python
from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(df['close'], model='additive', period=30)

decomposition.plot()
plt.show()
```

#### Final Interpretation (Ready for PPT)

You can use this:

The seasonal component shows minimal repeating patterns, indicating the absence of strong seasonality.
Most variations in the stock price are explained by the long-term trend, while residual fluctuations represent short-term market volatility.

#### Moving Average Smoothing

```python
df['MA_30'] = df['close'].rolling(window=30).mean()

plt.figure(figsize=(12,5))
plt.plot(df['close'], label='Original')
plt.plot(df['MA_30'], label='30-Day Moving Average')
plt.legend()
plt.title("Stock Price with Moving Average")
plt.show()
```

#### 
The 30-day moving average smooths short-term price fluctuations and highlights the overall upward trend in stock prices.
While daily volatility is present, the long-term direction remains consistently positive, indicating steady growth over time.
